```
from six import BytesIO
from PIL import Image
import numpy as np
import os

import matplotlib
import matplotlib.pyplot as plt
import glob
```

```
def load_image_into_numpy(path):
  img = open(path, 'rb').read()
  image = Image.open(BytesIO(img))
  (width, height) = image.size
  return np.array(image.getdata()).reshape((height, width, 3)).astype(np.uint8)
```

```
airplane_path = "/content/downloads/airplanes/*"
air_img_np = []
for i in glob.glob(airplane_path):
  air_img_np.append(load_image_into_numpy(i))
```

```
plt.rcParams['figure.figsize'] = [14, 7]

for idx, air_img_np in enumerate(air_img_np):
  plt.subplot(2, 3, idx+1)
  plt.imshow(air_img_np)
plt.show()  
```

```
heli_path = "/content/downloads/helicopters/*"
heli_img_np = []
for i in glob.glob(heli_path):
  heli_img_np.append(load_image_into_numpy(i))
```

```
plt.rcParams['figure.figsize'] = [14, 7]

for idx, heli_img_np in enumerate(heli_img_np):
  plt.subplot(2, 3, idx+1)
  plt.imshow(heli_img_np)
plt.show()  
```
